# Spectral Analysis Pipeline


```python
from qk_spectral_analysis import (
    generate_cluster_figures,
    generate_family_spectra,
    generate_stat_boxplots,
)


def display_cluster_progress(current: int, total: int, output_path):
    print(f"[{current}/{total}] Saved {output_path}")

FAMILY = "Qwen3"
DATA_ROOT = "."
FORCE_DOWNLOAD = False
N_CLUSTERS = 6
N_BINS = 24
```

```python
generated = generate_family_spectra(
    FAMILY, output_root=DATA_ROOT, force=FORCE_DOWNLOAD
)
print(f"Generated {len(generated)} new spectra files")
for path in generated:
    print(f" - {path}")
```

```python
cluster_figs = generate_cluster_figures(
    FAMILY,
    n_clusters=N_CLUSTERS,
    n_bins=N_BINS,
    data_root=DATA_ROOT,
    progress=display_cluster_progress,
)
print(f"Created {len(cluster_figs)} cluster figures")
```

```python
boxplots = generate_stat_boxplots(DATA_ROOT, family=FAMILY)
total_boxplots = len(boxplots)
print(f"Created {total_boxplots} statistical boxplots")
for idx, path in enumerate(boxplots, start=1):
    print(f"[{idx}/{total_boxplots}] {path}")
```
